Task 1

Define the DNA sequences and their IDs

```python
# read sequences
dna_seq = {} # creat a dictionary 
with open("dna_raw.txt", "r") as file: 
    seq_name = ""
    for line in file:
        line = line.strip() # we remove spaces and newline
        if line.startswith(">"):
            seq_name = line
            dna_seq [seq_name] = "" # start an empty sequence
        else:
            dna_seq[seq_name] += line # add to the sequence

```

print each DNA sequence together with their IDs

```python
# print each sequence name and dna sequence on separate lines
for seq_name, sequence in dna_seq.items():
    print (seq_name)
    print (sequence)
```

```python
# Counting bases
def count_base (dna_seq):
    counts = {"a":0, "t":0, "c":0, "g":0} # Initialize counts 
    for base in dna_seq.lower(): # convert to lower case to handla both cases
        if base in counts:  # check if base is one of a, t, g, c
            counts[base] += 1 # every time we see a base, increase its count by 1
    return counts # return the dictionary with counts
```

- Count how many times each DNA base (a, t, c, g) appears in a given DNA sequence
- returns the counts as a dictionary (to hold counts of each nucleotide) 

```python
# Count bases in each sequence and print the results

for seq_name, seq in dna_seq.items():
    counts = count_base(seq)
    print (f"The final counts {seq_name} : {counts}")
```
